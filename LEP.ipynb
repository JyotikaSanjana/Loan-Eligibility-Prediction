```python
df[['ApplicantIncome','LoanAmount']]
# Handle Numerical Missing Data
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mean())
```

```python
#Handling categorical Missing Data
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df
```

```python
#Label Encoding
df.Gender=df.Gender.map({'Male':1,'Female':0})
df.Gender.value_counts()
```

```python
df.Self_Employed=df.Self_Employed.map({'Yes':1,'No':0})
df.Self_Employed.value_counts()
```

```python
df.Married=df.Married.map({'Yes':1,'No':0})
df.Married.value_counts()
```

```python
df.Education=df.Education.map({'Graduate':1,'Not Graduate':0})
df.Education.value_counts()
```

```python
df.Dependents=df.Dependents.map({'0':0,'1':1,'2':2,'3+':3})
df.Dependents.value_counts()
```

```python
df.Property_Area=df.Property_Area.map({'Urban':0,'Rural':1,'Semiurban':2})
df.Property_Area.value_counts()
```

```python
df
```

```python
#Separating dependent and Independent variables
x = df.drop(columns=["Loan_ID",'Loan_Status'], axis=1)
y = df['Loan_Status']
```

```python
#Splitting Dataset into train and test Datasets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
x_train=sc_X.fit_transform(x_train)
x_test=sc_X.transform(x_test)
```

```python
#RandomForest Model
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(n_estimators=5,criterion="entropy",random_state=100)
model1.fit(x_train,y_train)

print("Accuracy of RFClassifier:", model1.score(x_test, y_test)*100)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier(criterion="entropy",random_state=0)
model2.fit(x_train, y_train)

print("Accuracy of Decision Tree:", model2.score(x_test, y_test)*100)
```

```python
#Logistic regression
from sklearn.linear_model import LogisticRegression
model3 = LogisticRegression(random_state=0)
model3.fit(x_train, y_train)

print("Accuracy of Logistic Regression:", model3.score(x_test, y_test)*100)
```

```python
#SVC
from sklearn.svm import SVC
model4 = SVC(C=5)
model4.fit(x_train, y_train)


print("Accuracy of SVC:", model4.score(x_test, y_test)*100)
```

```python
#KNN
from sklearn.neighbors import (KNeighborsClassifier)
model5=KNeighborsClassifier(n_neighbors=9, p=2,metric='euclidean')
model5.fit(x_train,y_train)

print("Accuracy of KNN:", model5.score(x_test, y_test)*100)
```

```python
#All Accuracies
print("Accuracy of RFClassifier:", model1.score(x_test, y_test)*100)
print("Accuracy of Decision Tree:", model2.score(x_test, y_test)*100)
print("Accuracy of LR:", model3.score(x_test, y_test)*100)
print("Accuracy of SVC:", model4.score(x_test, y_test)*100)
print("Accuracy of KNN:", model5.score(x_test, y_test)*100)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix,classification_report
y_pred = model3.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
cm

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True)
```
